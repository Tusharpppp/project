```python
def add_student_marks(students, name, subject1_marks, subject2_marks, subject3_marks):
    students[name] = {"Physics": subject1_marks, "chemistry": subject2_marks, "Maths": subject3_marks}

def display_all_students(students):
    if not students:
        print("No student records found. Add marks first.")
    else:
        for name, marks in students.items():
            print(f"{name}'s Marks: {marks}")

def search_student_info(students, name):
    if name in students:
        return students[name]
    else:
        return None

def delete_student(students, name):
    if name in students:
        del students[name]
        print(f"{name} removed from the records successfully!")
    else:
        print(f"{name} not found. No student removed.")

def main():
    students = {}

    while True:
        print("1. Add Student Marks")
        print("2. Display All Student Marks")
        print("3. Search Student Info")
        print("4. Delete Student")
        print("5. Quit")

        choice = input("Enter your choice (1/2/3/4/5): ")

        if choice == '1':
            name = input("Enter student name: ")
            subject1_marks = int(input("Enter marks for Physics: "))
            subject2_marks = int(input("Enter marks for Chemistry: "))
            subject3_marks = int(input("Enter marks for Maths: "))
            add_student_marks(students, name, subject1_marks, subject2_marks, subject3_marks)
            print("Marks added successfully!")

        elif choice == '2':
            display_all_students(students)

        elif choice == '3':
            name = input("Enter student name to search info: ")
            student_info = search_student_info(students, name)
            if student_info:
                print(f"{name}'s Info: {student_info}")
            else:
                print(f"{name} not found. No student info to display.")

        elif choice == '4':
            name = input("Enter student name to delete: ")
            delete_student(students, name)

        elif choice == '5':
            print("Exiting the program. Goodbye!")
            break

        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
